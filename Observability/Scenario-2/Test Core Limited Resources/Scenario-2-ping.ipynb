```python
df_low_logs = pd.read_csv('csv/low/logs.csv',  delimiter='::', header=0, engine='python')
df_low_logs['time'] = pd.to_datetime(df_low_logs['time'])
df_low_logs_count = df_low_logs.groupby(['time','kubernetes.container_name']).size().reset_index(name='counts')
df_low_logs_count_core = df_low_logs_count.loc[df_low_logs_count['kubernetes.container_name'] == 'free5gc']
df_low_logs_count_cu = df_low_logs_count.loc[df_low_logs_count['kubernetes.container_name'] == 'cu']
df_low_logs_count_du = df_low_logs_count.loc[df_low_logs_count['kubernetes.container_name'] == 'du']
df_low_logs_count_ru = df_low_logs_count.loc[df_low_logs_count['kubernetes.container_name'] == 'ru']

df_high_logs = pd.read_csv('csv/high/logs.csv',  delimiter='::', header=0, engine='python')
df_high_logs['time'] = pd.to_datetime(df_high_logs['time'])
df_high_logs_count = df_high_logs.groupby(['time','kubernetes.container_name']).size().reset_index(name='counts')
df_high_logs_count_core = df_high_logs_count.loc[df_high_logs_count['kubernetes.container_name'] == 'free5gc']
df_high_logs_count_cu = df_high_logs_count.loc[df_high_logs_count['kubernetes.container_name'] == 'cu']
df_high_logs_count_du = df_high_logs_count.loc[df_high_logs_count['kubernetes.container_name'] == 'du']
df_high_logs_count_ru = df_high_logs_count.loc[df_high_logs_count['kubernetes.container_name'] == 'ru']
```

```python
all_seconds_low = pd.date_range('2022-01-27 14:30:00', '2022-01-27 14:37:00', freq='s')
all_seconds_high = pd.date_range('2022-01-27 14:22:00', '2022-01-27 14:29:00', freq='s')

df_low_logs_count_cu.index = pd.DatetimeIndex(df_low_logs_count_cu['time'])
df_low_logs_count_cu = df_low_logs_count_cu.reindex(all_seconds_low)
df_low_logs_count_cu['kubernetes.container_name'] = df_low_logs_count_cu['kubernetes.container_name'].fillna('cu')
df_low_logs_count_cu['counts'] = df_low_logs_count_cu['counts'].fillna(0)

df_high_logs_count_cu.index = pd.DatetimeIndex(df_high_logs_count_cu['time'])
df_high_logs_count_cu = df_high_logs_count_cu.reindex(all_seconds_high)
df_high_logs_count_cu['kubernetes.container_name'] = df_high_logs_count_cu['kubernetes.container_name'].fillna('cu')
df_high_logs_count_cu['counts'] = df_high_logs_count_cu['counts'].fillna(0)

df_low_logs_count_du.index = pd.DatetimeIndex(df_low_logs_count_du['time'])
df_low_logs_count_du = df_low_logs_count_du.reindex(all_seconds_low)
df_low_logs_count_du['kubernetes.container_name'] = df_low_logs_count_du['kubernetes.container_name'].fillna('du')
df_low_logs_count_du['counts'] = df_low_logs_count_du['counts'].fillna(0)

df_high_logs_count_du.index = pd.DatetimeIndex(df_high_logs_count_du['time'])
df_high_logs_count_du = df_high_logs_count_du.reindex(all_seconds_high)
df_high_logs_count_du['kubernetes.container_name'] = df_high_logs_count_du['kubernetes.container_name'].fillna('du')
df_high_logs_count_du['counts'] = df_high_logs_count_du['counts'].fillna(0)

df_low_logs_count_ru.index = pd.DatetimeIndex(df_low_logs_count_ru['time'])
df_low_logs_count_ru = df_low_logs_count_ru.reindex(all_seconds_low)
df_low_logs_count_ru['kubernetes.container_name'] = df_low_logs_count_ru['kubernetes.container_name'].fillna('ru')
df_low_logs_count_ru['counts'] = df_low_logs_count_ru['counts'].fillna(0)

df_high_logs_count_ru.index = pd.DatetimeIndex(df_high_logs_count_ru['time'])
df_high_logs_count_ru = df_high_logs_count_ru.reindex(all_seconds_high)
df_high_logs_count_ru['kubernetes.container_name'] = df_high_logs_count_ru['kubernetes.container_name'].fillna('ru')
df_high_logs_count_ru['counts'] = df_high_logs_count_ru['counts'].fillna(0)

df_low_logs_count_core.index = pd.DatetimeIndex(df_low_logs_count_core['time'])
df_low_logs_count_core = df_low_logs_count_core.reindex(all_seconds_low)
df_low_logs_count_core['kubernetes.container_name'] = df_low_logs_count_core['kubernetes.container_name'].fillna('free5gc')
df_low_logs_count_core['counts'] = df_low_logs_count_core['counts'].fillna(value=0)

df_high_logs_count_core.index = pd.DatetimeIndex(df_high_logs_count_core['time'])
df_high_logs_count_core = df_high_logs_count_core.reindex(all_seconds_high)
df_high_logs_count_core['kubernetes.container_name'] = df_high_logs_count_core['kubernetes.container_name'].fillna('free5gc')
df_high_logs_count_core['counts'] = df_high_logs_count_core['counts'].fillna(value=0)
```

```python
df_low_logs_count_sum = pd.DataFrame()
df_low_logs_count_sum = df_low_logs_count_sum.reindex(all_seconds_low)
df_low_logs_count_sum['ru'] = df_low_logs_count_ru['counts']
df_low_logs_count_sum['ru-du'] = df_low_logs_count_ru['counts'] + df_low_logs_count_du['counts']
df_low_logs_count_sum['ru-du-cu'] = df_low_logs_count_ru['counts'] + df_low_logs_count_du['counts'] + df_low_logs_count_cu['counts']

df_high_logs_count_sum = pd.DataFrame()
df_high_logs_count_sum = df_high_logs_count_sum.reindex(all_seconds_high)
df_high_logs_count_sum['ru'] = df_high_logs_count_ru['counts']
df_high_logs_count_sum['ru-du'] = df_high_logs_count_ru['counts'] + df_high_logs_count_du['counts']
df_high_logs_count_sum['ru-du-cu'] = df_high_logs_count_ru['counts'] + df_high_logs_count_du['counts'] + df_high_logs_count_cu['counts']
```
